### 1) Print a square pattern of stars of size n. Input n from the user. Example for n=4:
```
****
****
****
****
```

**Approach:** Use a nested loop: outer loop for rows (0..n-1) and inner loop to print n stars in each row. Use print(..., end='') and print() to move to next line.


````python
# Write your code here
n=int(input("enter the value of n: "))
for i in range(n):
  for j in range(n):
    print("*",end="")
  print()


````

### 4) Print a hollow square of size n. Example for n=5:
```
*****
*   *
*   *
*   *
*****
```

**Approach:** Print full stars for first and last row. For middle rows, print star, (n-2) spaces, star. Use loops and conditionals.


````python
# Write your code here
n=int(input("enter the value of n: "))
for i in range(n):
  for j in range(n):
    if i==0 or i==n-1 or j==0 or j==n-1:
      print("*",end="")
    else:
      print(" ",end="")
  print()

````

### 10) Given n, print a checkerboard of size n x n using X and O alternately. Example n=4:

**Approach:** For position (i,j), if (i+j) is even print 'X' else 'O'. Use nested loops and conditionals.

### 19) Given n, print a symmetric diamond pattern of stars with maximum width 2*n-1. Example n=3:
```
  *
 ***
*****
 ***
  *
```

**Approach:** Use two parts: top (1..n) and bottom (n-1..1). For row k compute spaces and stars accordingly.


````python

# Write your code here
n=int(input("enter the value of n: "))

for i in range(n-1):
  for j in range(i,n):
    print(" ",end="")
  for j in range(i):
    print("*",end="")
  for j in range(i+1):
    print("*",end="")
  print()
for i in range(n):
  for j in range(i+1):
    print(" ",end="")
  for j in range(i,n-1):
    print("*",end="")
  for j in range(i,n):
    print("*",end="")
  print()


````

### 20) Given n, print an n x n matrix where rows alternate between increasing numbers 1..n and decreasing numbers n..1. Example n=3:
```
1 2 3
3 2 1
1 2 3
```

**Approach:** For each row i, if i is odd print 1..n else print n..1. Use nested loops and conditionals.


````python
# Write your code here
n=int(input("enter the value of n: "))
for i in range(1,n+1):
  if i%2!=0:
    for j in range(1,n+1):
      print(j,end=" ")
  else:
    for j in range(n,0,-1):
      print(j,end=" ")
  print()

````
